Training set: Avocado prices from Kaggle

```python
%matplotlib inline
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
avocado = pd.read_csv('avocado.csv')
avocado.drop(columns='Unnamed: 0', inplace=True)
avocado['Date'] = pd.to_datetime(avocado['Date'])
avocado.sort_values(by='Date', ascending=True, inplace=True)
avocado.columns = [x.upper() for x in avocado.columns]
avocado.columns = avocado.columns.str.replace(' ', '_')
avocado.head(10)
```

```python
plt1 = avocado.groupby('DATE')['TOTAL_VOLUME'].mean().reset_index()
bins = [x for x in range(200000, 1700001, 50000)]
avg = plt1['TOTAL_VOLUME'].mean()
median = plt1['TOTAL_VOLUME'].median()

plt.style.use('fivethirtyeight')
plt.hist(plt1['TOTAL_VOLUME'], bins=bins, color='#ACE1AF',
         edgecolor='black')
plt.axvline(avg, color='#211951', label='Avg Volume',
            linestyle='dashed', linewidth=2)
plt.axvline(median, color='#0B666A', label='Median Volume',
            linestyle='dotted', linewidth=2)


plt.ylabel('No. of weeks', fontsize=12)
plt.xlabel('Mean Total Volume', fontsize=12)

plt.legend(loc='upper right')
plt.title('Avocado Total Volume Distribution', fontsize=14)
plt.savefig('plots\\avocado_total_volume_distribution.png')
plt.show()
```
