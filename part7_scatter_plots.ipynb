# Scatter Plots

Corey Schafer's YT Matplotlib Tutorial: Part 7

Training set: Avocado prices from Kaggle

```python
%matplotlib inline
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import random
```

```python
avocado = pd.read_csv('avocado.csv')
avocado.drop(columns='Unnamed: 0', inplace=True)
avocado['Date'] = pd.to_datetime(avocado['Date'])
avocado.sort_values(by='Date', ascending=True, inplace=True)
avocado.columns = [x.upper() for x in avocado.columns]
avocado.columns = avocado.columns.str.replace(' ', '_')
avocado.head(10)
```

```python
plt1 = avocado.groupby('TOTAL_VOLUME')[['SMALL_BAGS']].mean().reset_index()
colours = [random.randint(0, 10) for x in plt1['TOTAL_VOLUME']]

plt.style.use('fivethirtyeight')
plt.scatter(plt1['TOTAL_VOLUME'], plt1['SMALL_BAGS'], s=5,
            c=colours, cmap='Greens', edgecolor='black',
            linewidth=0.1, alpha=0.8)

cbar = plt.colorbar()
cbar.set_label('Reliability Level')

plt.xscale('log')
plt.yscale('log')

plt.xlabel('Total Volume', fontsize=12)
plt.ylabel('Small Bags', fontsize=12)

plt.savefig('plots\\small_bags_vs_total_volume.png')
plt.title('Small Bags Usage vs. Avocado Total Volume', fontsize=16)
```
