# Stack Plots

Corey Schafer's YT Matplotlib Tutorial: Part 4

Training set: Avocado prices from Kaggle

```python
%matplotlib inline
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
avocado = pd.read_csv('avocado.csv')
avocado.drop(columns='Unnamed: 0', inplace=True)
avocado['Date'] = pd.to_datetime(avocado['Date'])
avocado.sort_values(by='Date', ascending=True, inplace=True)
avocado.columns = [x.upper() for x in avocado.columns]
avocado.columns = avocado.columns.str.replace(' ', '_')
avocado.head(10)
```

```python
plt1 = avocado.groupby('DATE')[['SMALL_BAGS', 'LARGE_BAGS', 'XLARGE_BAGS']].sum().reset_index()
labels = ['S bags', 'L bags', 'XL bags']

plt.style.use('fivethirtyeight')
plt.stackplot(plt1['DATE'], plt1['SMALL_BAGS'], plt1['LARGE_BAGS'], plt1['XLARGE_BAGS'], labels=labels)
plt.xticks(rotation=30, fontsize=10, horizontalalignment='right')
plt.title('AVOCADO BAG TYPE OVER TIME', fontsize=14)
plt.legend(loc='upper left')
plt.savefig('plots\\avocado_bag_type_over_time.png')
plt.show()
```
