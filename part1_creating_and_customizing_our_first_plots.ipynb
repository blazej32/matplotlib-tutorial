# Creating and Customizing Our First Plots

Corey Schafer's YT Matplotlib Tutorial: Part 1

Training set: Avocado prices from Kaggle

```python
%matplotlib inline
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
avocado = pd.read_csv('avocado.csv')
avocado.drop(columns='Unnamed: 0', inplace=True)
avocado['Date'] = pd.to_datetime(avocado['Date'])
avocado.sort_values(by='Date', ascending=True, inplace=True)
avocado.head(10)
```

```python
plt1 = avocado.groupby('Date')['AveragePrice'].mean().reset_index()
plt.plot(plt1['Date'], plt1['AveragePrice'])
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.title('Avocado Price Over Time')
plt.show()
```
