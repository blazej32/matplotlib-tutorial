# Pie Charts

Corey Schafer's YT Matplotlib Tutorial: Part 3

Training set: Avocado prices from Kaggle

```python
%matplotlib inline
import matplotlib as mpl
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
avocado = pd.read_csv('avocado.csv')
avocado.drop(columns='Unnamed: 0', inplace=True)
avocado['Date'] = pd.to_datetime(avocado['Date'])
avocado.sort_values(by='Date', ascending=True, inplace=True)
avocado.columns = [x.upper() for x in avocado.columns]
avocado.columns = avocado.columns.str.replace(' ', '_')
avocado.head(10)
```

```python
plt1 = avocado.groupby('TYPE')['TOTAL_VOLUME'].sum().reset_index()
explode = [0, 0.2]

plt.style.use('fivethirtyeight')
plt.pie(plt1['TOTAL_VOLUME'], labels=plt1['TYPE'], colors=['#953553', '#FFC300'],
         wedgeprops={'edgecolor': '#301934'}, explode=explode, shadow=True,
         startangle=0, autopct='%1.1f%%')

plt.title('TOTAL VOLUME PER AVOCADO TYPE', fontsize=14)
plt.savefig('plots\\total_volume_per_avocado_type.png')
plt.show()
```
